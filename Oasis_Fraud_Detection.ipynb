## Handling Imbalanced Dataset
Undersampling

```python
# separating the data for analysis
legit = new_df[new_df.Class == 0]
fraud = new_df[new_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample=legit.sample(n=473)
legit_sample.shape
```

```python
#creating new dataframe
new_df = pd.concat([legit_sample,fraud],ignore_index=True)
```

```python
new_df['Class'].value_counts()
```

```python
new_df.head()
```

```python
#feature variable and target variable
x = new_df.drop('Class',axis=1)  # axis=1 meansfull column will be dropped and axis = 0 will drop a row
y = new_df['Class']
```

```python
#spliting train and test set after balancing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,
                                                 random_state=42)
```

```python
#Logistic Regression
log = LogisticRegression()
log.fit(x_train,y_train)    # training the dataset using fit function ()
y_pred1 = log.predict(x_test)

print(f"accuracy score of Logistic Regression = {accuracy_score(y_test,y_pred1):.2f}")
```

```python
print("Logistic Regression\n")
print(f"precision score = {precision_score(y_test,y_pred1):.2f}")
print(f"recall score = {recall_score(y_test,y_pred1):.2f}")
print(f"f1 score = {f1_score(y_test,y_pred1):.2f}")
```

```python
#Decision Tree Classifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred2 = dt.predict(x_test)

print(f"accuracy score of Decision Tree Classifier = {accuracy_score(y_test,y_pred2):.2f}")
```

```python
print("Decision Tree Classifier\n")
print(f"precision score = {precision_score(y_test,y_pred2):.2f}")
print(f"recall score = {recall_score(y_test,y_pred2):.2f}")
print(f"f1 score = {f1_score(y_test,y_pred2):.2f}")
```

```python
#Random Forest Classifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred3 = rf.predict(x_test)

print(f"accuracy score of Random Forest Classifier = {accuracy_score(y_test,y_pred3):.2f}")
```

```python
print("Random Forest Classifier\n")
print(f"precision score = {precision_score(y_test,y_pred3):.2f}")
print(f"recall score = {recall_score(y_test,y_pred3):.2f}")
print(f"f1 score = {f1_score(y_test,y_pred3):.2f}")
```

```python
final_data = pd.DataFrame({'Models':['LR','DT','RF'],
              "Accuracy score":[accuracy_score(y_test,y_pred1)*100,
                     accuracy_score(y_test,y_pred2)*100,
                     accuracy_score(y_test,y_pred3)*100
                    ]})
final_data
```
